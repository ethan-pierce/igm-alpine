```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
optimize = Dataset('./optimize.nc')
geology = Dataset('./geology-optimized.nc')
```

```python
optimize.variables.keys()
```

```python
geology.variables.keys()
```

```python
im = plt.imshow(geology['thk'])
plt.title('Ice thickness (m)')
plt.colorbar(im)
plt.show()
```

```python
im = plt.imshow(geology['usurf'])
plt.title('Surface elevation (m)')
plt.colorbar(im)
plt.show()
```

```python
im = plt.imshow(optimize['velsurf_mag'][-1])
plt.title('Surface velocity (m/a)')
plt.colorbar(im)
plt.show()
```

```python
im = plt.imshow(geology['slidingco'])
plt.title('Sliding coefficient')
plt.colorbar(im)
plt.show()
```

```python
sliding_velocity = np.loadtxt('sliding_velocity_magnitude.txt')
im = plt.imshow(sliding_velocity)
plt.title('Sliding velocity (m/a)')
plt.colorbar(im)
plt.show()
```

```python
error = np.loadtxt('rms_std.dat')
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 3))

ax[0].plot(error[:,0])
ax[0].set_title('RMS ice thickness')

ax[1].plot(error[:,1])
ax[1].set_title('Std. ice thickness')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 3))

ax[0].plot(error[:,2])
ax[0].set_title('RMS ice velocity')

ax[1].plot(error[:,3])
ax[1].set_title('Std. ice velocity')

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 3))

ax[0].plot(error[:,6])
ax[0].set_title('RMS ice surface')

ax[1].plot(error[:,7])
ax[1].set_title('Std. ice surface')

plt.show()
```
