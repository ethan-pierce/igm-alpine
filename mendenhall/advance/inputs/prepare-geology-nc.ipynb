```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
from netCDF4 import Dataset
```

```python
with rio.open('bedrock.tif') as f:
    topg_data = f.read(1)
    cols, rows = np.meshgrid(np.arange(f.shape[1]), np.arange(f.shape[0]))
    xs, ys = rio.transform.xy(f.transform, rows, cols)
    lon_vals = np.array(xs)
    lat_vals = np.array(ys)
    
    thk_data = np.zeros_like(topg_data)
```

```python
geo = Dataset('geology.nc', 'r+', format = 'NETCDF4')
y = geo.createDimension('y', len(lat_vals[:,0]))
x = geo.createDimension('x', len(lon_vals[0,:]))

y = geo.createVariable('y', 'f4', ('y',))
x = geo.createVariable('x', 'f4', ('x',))
```

```python
topg = geo.createVariable('topg', 'f4', ('y', 'x',))
thk = geo.createVariable('thk', 'f4', ('y', 'x',))
```
