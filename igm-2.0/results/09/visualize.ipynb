```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
obs = Dataset('geology.nc')
print('obs: ', obs.variables.keys())

try:
    geol = Dataset('geology-optimized.nc')
    print('geol :', geol.variables.keys())
except:
    pass
```

```python
valid = geol['thk'][:] > 0.5
field = (geol['thk'][:])
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = (geol['slidingco'][:])
print(np.mean(field))
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = geol['velsurf_mag'][:] - geol['velsurfobs_mag'][:]
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = geol['thk'][:] - obs['thkobs'][:]
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = np.loadtxt('sliding_velocity_mag.txt')
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```
