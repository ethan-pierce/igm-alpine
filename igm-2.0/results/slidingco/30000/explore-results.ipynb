```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
geo = Dataset('geology-optimized.nc')
print(geo.variables.keys())
      
ds = Dataset('output.nc')
print(ds.variables.keys())
```

```python
field = ds.variables['velbase_mag'][-1]
field = np.flip(field * geo.variables['icemask'], axis = 0)
im = plt.imshow(field, cmap = 'jet')
plt.colorbar(im)
plt.show()
```

```python
field = ds.variables['velsurf_mag'][-1]
field = np.flip(field * geo.variables['icemask'], axis = 0)
im = plt.imshow(field, cmap = 'jet')
plt.colorbar(im)
plt.show()
```

```python
field = geo.variables['velsurfobs_mag'][:]
field = np.flip(field * geo.variables['icemask'], axis = 0)
im = plt.imshow(field, cmap = 'jet')
plt.colorbar(im)
plt.show()
```

```python
field = np.abs((ds.variables['velsurf_mag'][-1] - geo.variables['velsurfobs_mag'][:]) / geo.variables['velsurfobs_mag'][:]) * 100
field = np.flip(field * geo.variables['icemask'], axis = 0)
im = plt.imshow(field, vmax = 200, cmap = 'jet')
plt.colorbar(im)
plt.show()
```
