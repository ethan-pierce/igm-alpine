```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
geo = Dataset('geology-optimized.nc')
print(geo.variables.keys())
      
ds = Dataset('output.nc')
print(ds.variables.keys())
```

```python
mask = ds.variables['thk'][-1] > 0.5
field = np.flip(np.where(mask, ds.variables['velbase_mag'][-1], np.nan), axis = 0)
im = plt.imshow(field)
plt.colorbar(im)
plt.show()
```
