```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
data = Dataset('optimize.nc')
print('data: ', data.variables.keys(), data['thk'].shape)

try:
    geol = Dataset('geology-optimized.nc')
    print('geol :', geol.variables.keys())
except:
    pass
```

```python
valid = data['thk'][-1] > 0.5
field = (data['thk'][-1])
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = data['thk'][-1] > 0.5
field = (data['slidingco'][-1])
print(np.mean(field))
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = data['thk'][-1] > 0.5
field = (data['velsurf_mag'][-1] - data['velsurfobs_mag'][-1])
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot, vmax = 100, vmin = -100)
plt.colorbar(im)
plt.show()
```
