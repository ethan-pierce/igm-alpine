```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
obs = Dataset('geology.nc')
print('obs: ', obs.variables.keys())

try:
    geol = Dataset('geology-optimized.nc')
    print('geol :', geol.variables.keys())
except:
    pass

try:
    opti = Dataset('optimize.nc')
    print('opti :', opti.variables.keys())
except:
    pass
```

```python
valid = geol['thk'][:] > 0.5
field = (geol['thk'][:])
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = (opti['slidingco'][14])
print(np.mean(field))
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
t = 6
valid = geol['thk'][:] > 0.5
field = opti['velsurf_mag'][t] - opti['velsurfobs_mag'][t]
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = geol['thk'][:] - obs['thkobs'][:]
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = np.loadtxt('sliding_velocity_mag.txt')
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```
