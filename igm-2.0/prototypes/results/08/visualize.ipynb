```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
```

```python
# data = Dataset('optimize.nc')
# print('data: ', data.variables.keys(), data['thk'].shape)

try:
    geol = Dataset('geology-optimized.nc')
    print('geol :', geol.variables.keys())
except:
    pass
```

```python
valid = geol['thk'][:] > 0.5
field = (geol['thk'][:])
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = (geol['slidingco'][:])
print(np.mean(field))
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot)
plt.colorbar(im)
plt.show()
```

```python
valid = geol['thk'][:] > 0.5
field = np.abs((geol['velsurf_mag'][:] - geol['velsurfobs_mag'][:]) / geol['velsurfobs_mag'][:]) * 100
toplot = np.where(valid, field, np.nan)
im = plt.imshow(toplot, vmax = 60, vmin = 0)
plt.colorbar(im)
plt.show()
```
